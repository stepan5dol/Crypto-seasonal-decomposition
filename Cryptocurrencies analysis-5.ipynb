```python
from statsmodels.tsa.seasonal import seasonal_decompose
import pandas as pd
import pandas
import matplotlib.pyplot as plt
import matplotlib
import seaborn as sns
```

```python
def openfunc(filename):
    table = pd.read_csv(filename, sep=';', index_col=[6], parse_dates=[6])
    display(table.head())
    print(f'Is data in chronological order? {table.index.is_monotonic}')
    return table
```

## Bitcoin analysis

```python
bit = openfunc('/Users/stepan5dol/Downloads/BTC_ALL_graph_coinmarketcap.csv')
bit = bit.drop(columns=['volume', 'marketCap'])['2017':'2023']
    
plt.figure(figsize=(15,8))
sns.lineplot(data=bit, palette="husl")
plt.legend(bit.columns)
plt.xlabel('Time')
plt.ylabel('Price')
plt.title('Fig.1: Bitcoin price line graph for all time')
plt.show()
```

```python
bit_dec = bit.resample('1M').median()

plt.figure(figsize=(15,8))
sns.lineplot(data=bit_dec, palette="husl")
plt.legend(bit.columns)
plt.xlabel('Time')
plt.ylabel('Price')
plt.title('Fig.1.1: Bitcoin price resampled 1M period price line graph for all time')
plt.show()
```
